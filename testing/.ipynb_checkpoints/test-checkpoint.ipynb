```python
from pySPDC import crystals as c
from pySPDC import utils as u
import numpy as np
import scipy.constants as constants
import numba
import matplotlib.pyplot as plt
import pandas as pd
from seaborn import color_palette
```

```python
shift = 0
p_wl = 404.9
p_wl = 775
pump_wl   = p_wl*(1e-9)
signal_wl = ((2*p_wl)+shift)*(1e-9)
idler_wl  = ((2*p_wl)-shift)*(1e-9)
print(pump_wl, signal_wl, idler_wl)
ktp = c.KTP(pump_wl, signal_wl, idler_wl,
            axes={'pump':'y', 'signal':'y', 'idler':'z'})
```

```python
temperature = 25
length_at_25 = 1e-2 #crystal length spec at 25
crystal_L = length_at_25*(1+6.7e-6*(temperature-25)+11e-9*(temperature-25))

omega_p = 2*np.pi*constants.c/pump_wl
omega_c = 2*np.pi*constants.c/signal_wl

fwhm_p = 1.30e-9
fwhm_p = 0.65e-9
fwhm_p = 0.5e-9
sigma = 2*np.pi*u.bandwidth_conversion(fwhm_p,pump_wl)

grid_size = 501
grid_size = 256

sf = np.array([0,-1])
pm = np.array([-1,1])
spectral_window = 12.5e-9 
spectral_window = 6.25e-9 
spectral_window = 5e-9 
idler_range = (2*np.pi*constants.c)/np.linspace(*(idler_wl + (pm*spectral_window)),grid_size)
signal_range = (2*np.pi*constants.c)/np.linspace(*(signal_wl + (pm*spectral_window)),grid_size)

print(idler_range[sf],signal_range[sf])
```

```python
poling_period = ktp.calc_poling_period(wl_p=ktp.pump_wavelength,
                                  wl_s=ktp.signal_wavelength,
                                 wl_i=ktp.idler_wavelength)
print(poling_period)
```

```python
dk = ktp.phase_matching_matrix(signal_range, idler_range, temperature)
```

```python
print(dk)
```

```python
pmf = np.sinc((dk - ((2*np.pi) / poling_period)) * (crystal_L/2))
print(pmf.shape)

#pmf = u.phase_matching_function_gaussian(dk, poling_period, crystal_L/2)
pmf = u.phase_matching_function_antisymmetric(dk, poling_period, crystal_L/2)

pe = u.calc_pump_envelope(signal_range,idler_range,omega_p,sigma)

jsa = pmf*pe
jsi = jsa*np.conjugate(jsa)
```

```python
pe.min(), pe.max()
pe[pe == pe.min()] = 0
```

```python
nc, nr = 2, 2
fig, axes = plt.subplots(ncols=nc,nrows=nr,figsize=[8,8],sharex=True,sharey=True,squeeze=True)
colours = color_palette("flare_r", as_cmap=True)
colours = color_palette("icefire", as_cmap=True)
axes[0][0].pcolormesh(dk, shading="gouraud",cmap=colours)
axes[0][1].pcolormesh(pe, shading="gouraud",cmap=colours)
axes[1][0].pcolormesh(np.real(pmf.T), shading="gouraud", cmap=colours)
# axes[1][1].pcolormesh(np.sqrt(jsi.T), shading="gouraud", cmap=colours)
axes[1][1].pcolormesh(jsa.T.real, shading="gouraud", cmap=colours)

txt_x = 0.05*dk.shape[0]
txt_y = 0.05*dk.shape[1]
txt_pos = (txt_x,txt_y)

fs = 20

axes[0][0].text(*txt_pos, r"$\Delta k$", fontsize=fs, color="w")
axes[0][1].text(*txt_pos, r"$\alpha\left(\omega_{s},\omega_{i}\right)$", fontsize=fs, color="w")
axes[1][0].text(*txt_pos, r"$\phi\left(\omega_{s},\omega_{i}\right)$", fontsize=fs, color="w")
axes[1][1].text(*txt_pos, r"$\alpha\left(\omega_{s},\omega_{i}\right)\phi\left(\omega_{s},\omega_{i}\right)$", fontsize=fs, color="w")

n_labels = 5
idler_pos = np.linspace(0,idler_range.shape[0]-1,n_labels,dtype=int)
idler_labels = ((2*np.pi*constants.c) / idler_range[idler_pos])*(1e9)
idler_labels = ["%.2f"%wl for wl in idler_labels]

signal_pos = np.linspace(0,signal_range.shape[0]-1,n_labels,dtype=int)
signal_labels = ((2*np.pi*constants.c) / signal_range[signal_pos])*(1e9)
signal_labels = ["%.2f"%wl for wl in signal_labels]


for col in range(nc):
    for row in range(nr):
        
        if row == 1:
            rotation = "vertical"
        else:
            rotation = None
            
        axes[row][col].set_xticks(idler_pos)
        axes[row][col].set_xticklabels(idler_labels,rotation=rotation)
        axes[row][col].set_yticks(signal_pos)
        axes[row][col].set_yticklabels(signal_labels)
        axes[row][col].grid(alpha=0.6)
        
fig.suptitle(r"""$\omega_{p}=%.2f\mathrm{nm}$, $\omega_{s}=%.2f\mathrm{nm}$, $\omega_{i}=%.2f\mathrm{nm}$, $\Lambda=%.2f\mathrm{\mu m}$
Directions : [$\mathrm{pump_{%s}}$, $\mathrm{signal_{%s}}$, $\mathrm{idler_{%s}}$]"""%(
    pump_wl*(1e9),
    signal_wl*(1e9),
    idler_wl*(1e9),
    poling_period*(1e6),
    ktp.pump._get_axis(),
    ktp.signal._get_axis(),
    ktp.idler._get_axis()))
plt.tight_layout()
```

```python
def hong_ou_mandel_interference(joint_spectral_amplitude,
                                frequencies_signal,
                                frequencies_idler,
                                delays):
    
    u, d, v = np.linalg.svd(joint_spectral_amplitude)
    d_diag = np.diag(d)
    u_conj = np.conj(u)

    probabilities = np.zeros(delays.size)

    for i, tau in enumerate(delays):
        intereference_signal = np.dot(
            v * np.exp(-1j * frequencies_signal * tau),
            u_conj
        )
        
        intereference_idler = np.dot(
            u.T,
            np.conj(v * np.exp(-1j * frequencies_idler * tau)).T
        )

        interference = intereference_signal * intereference_idler
        interference = np.dot(d_diag, np.dot(interference, d_diag))

        probabilities[i] = interference.sum()

    probabilities = probabilities / probabilities.max()

    probabilities = 0.5 - (0.5 * probabilities)

    d = d / np.sqrt(np.sum(d * np.conj(d)))
    purity = np.sum((d * np.conj(d))**2)
    schmidt_number = 1 / np.sum(d**4)
    entropy = -np.sum((d**2) * np.log2(d**2))

    return probabilities, purity, schmidt_number, entropy
```

```python
tau = np.linspace(-10, 10, 256) * 1e-12
new_hom, purity, schmidt_number, entropy = hong_ou_mandel_interference(jsa.real, signal_range, idler_range, tau)
```

```python
fig = plt.figure()
plt.plot(tau, new_hom)
plt.title(f"Purity:{purity:.2f} | Schmidt Number:{schmidt_number:.2f} | Entropy:{entropy:.2f}")
```
